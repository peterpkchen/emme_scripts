```python
import os
import pandas as pd
```

```python
mat_pth = r"C:\Users\pechen\Desktop\OntarioLine_Modelling"
zone_pth = r"C:\Users\pechen\Documents\407 Model Update"
zone_map = pd.read_csv(os.path.join(zone_pth,"407_ZoneMapping_spicy.csv"))
```

```python
mat_test = pd.read_csv(os.path.join(mat_pth,"mf35.csv"))
```

```python
mat_test.head()
```

```python
mat_test2 = mat_test.melt(id_vars=['p/q/[val]'], value_vars=mat_test.columns.values.tolist()[1:], 
                    var_name='Dest', value_name='Trips').rename(columns = {'p/q/[val]': 'Ori'})
```

```python
mat_test2.head()
```

```python
mat_test2['Dest'] = pd.to_numeric(mat_test2['Dest'])
mat_test3 = mat_test2
mat_test3 = mat_test3.merge(zone_map, left_on = 'Ori', right_on = 'GTA06').drop(columns = ['GTA06', 'region', 'pd']).rename(columns = {'phf_group':"O_subregion"})
mat_test3 = mat_test3.merge(zone_map, left_on = 'Dest', right_on = 'GTA06').drop(columns = ['GTA06', 'region', 'pd']).rename(columns = {'phf_group':"D_subregion"})

```

```python
mat_test3
```

```python
mat_test4 = pd.pivot_table(mat_test3, index = ["O_subregion","D_subregion"], values = 'Trips', aggfunc = sum,margins = True).fillna(0)
```

```python
mat_test4
```

```python
mat_test4.to_csv(os.path.join(mat_pth,'yolo_swag.csv'))
```
