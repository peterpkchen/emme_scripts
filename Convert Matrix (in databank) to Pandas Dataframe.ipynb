```python
import os
import pandas as pd
import inro.modeller as _m

# Emme tools
mm = _m.Modeller()
tmg_import = 'tmg.input_output.import_binary_matrix'
import_mat = mm.tool(tmg_import)
inro_change_mat = 'inro.emme.data.matrix.change_matrix_properties'
change_mat = _m.Modeller().tool(inro_change_mat)
eb = _m.Modeller().emmebank

# matrix to pandas 
def mat_to_pddf(id, scen):
    mat = eb.matrix(id).get_data(scen)
    zones_tupl = mat.indices
    indices = pd.Index(zones_tupl[0])
    indices.name = 'p'
    cols = pd.Index(zones_tupl[1])
    cols.name = 'q'
    matrix = mat.to_numpy()
    return pd.DataFrame(matrix, index=indices, columns=cols)
```

### Change below as needed for analysis

```python
ml_transztsa = mat_to_pddf('mf3', 9999)
```
